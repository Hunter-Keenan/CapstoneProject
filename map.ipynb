```python
#charge.to_csv('train.csv', index=False)
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
trains = pd.read_csv('discharge_update1.csv')
```

```python
print(trains.head())
trains.dtypes
```

```python
trains['Capacity'].isnull().sum()
```

```python
# plot a heatmap to display +ve and -ve correlation among features and regression label:
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
featurs=['voltage_battery','current_battery','temp_battery','current_load','voltage_load','time']
import seaborn as sns
cm = np.corrcoef(trains[featurs].values.T)
sns.set(font_scale=1.0)
fig = plt.figure(figsize=(15, 10))
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 8}, yticklabels=featurs, xticklabels=featurs)
plt.title('Features Correlation Heatmap')
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline  

from pandas.tools.plotting import scatter_matrix
sns.reset_orig()
scatter_matrix(trains, alpha=0.2, figsize=(20, 20), diagonal='kde')
```
